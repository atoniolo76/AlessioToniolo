```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import random

# Set configuration for the plot
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

# Plot the figure in the console
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

depth = 20
arr = np.zeros((depth, depth, depth))

for x in range(len(arr[:, 0, 0])):
    for y in range(len(arr[0, :, 0])):
        for z in range(len(arr[0, 0, :])):
            arr[x, y, z] = 1
            rand_num = random.randint(4)

            add_num = False
            if arr[x, y, z-1] == 1 or arr[x-1, y, z-1] == 1 or arr[x+1, y, z-1] == 1:
              add_num = True

            if (rand_num == 1 and add_num) or z < depth / 3:
              ax.scatter(x, y, z, c="black")
              arr[x, y, z] = 1

# Display the data
plt.show()
```
